# Global Performance, overall reports (evaluating the charge possible)

```python
# We say that overall performance is the daily number of reports completed by an agency for a given day
```

```python
work_df = h_a.merge(report, how = 'left', on = 'Report ID')
```

```python
overall_perf = work_df.groupby(['Agency (Manual Processing Organisation)','Report Completed Date'])\
                      .agg({'Report ID': 'nunique'})
```

```python
total_tasks = work_df.groupby('Report Completed Date').agg({'Report ID': 'nunique'})
```

```python
overall_perf_ratio = overall_perf.reset_index().set_index('Report Completed Date')\
                        .merge(total_tasks, 
                               how = 'inner', on = 'Report Completed Date')
```

```python
overall_perf_ratio['Completed Ratio'] = (overall_perf_ratio['Report ID_x']/overall_perf_ratio['Report ID_y'])*100
```

```python
overall_perf_ratio = overall_perf_ratio \
                        .reset_index() \
                        .set_index(['Agency (Manual Processing Organisation)','Report Completed Date']) \
                        .rename(columns = {'Report ID_x': 'Number of daily reports', 
                                           'Report ID_y': 'Total Number of reports'})
```

```python
overall_perf_ratio.head(2)
```

```python
fig, ax = plt.subplots(2,1,figsize = (20,20))

ax[0].set_title('Performance based on number of reports for each agency (Ratio)')
sns.lineplot(x = 'Report Completed Date', y = 'Completed Ratio', 
                     data = overall_perf_ratio, 
                     hue = 'Agency (Manual Processing Organisation)', ax = ax[0])


ax[1].set_title('Performance based on number of reports for each agency (Number)')
sns.lineplot(x = 'Report Completed Date', y = 'Number of daily reports', 
                     data = overall_perf_ratio, 
                     hue = 'Agency (Manual Processing Organisation)', ax = ax[1])

```

# Performance for each type of tasks

```python
# I noticed that sometimes, tasks are split between several agencies.
# Maybe some optimizations to do about it?
# Can we try to see why?
```

```python
count_df = h_a.groupby('Report ID').agg({'Agency (Manual Processing Organisation)':'nunique'})
nb_several = count_df[count_df['Agency (Manual Processing Organisation)'] > 1].reset_index()['Report ID'].nunique()
nb_unique = count_df[~(count_df['Agency (Manual Processing Organisation)'] > 1)].reset_index()['Report ID'].nunique()

several_agencies_tasks = count_df[count_df['Agency (Manual Processing Organisation)'] > 1] \
                            .reset_index()['Report ID'].values

unique_agencies_tasks = count_df[~(count_df['Agency (Manual Processing Organisation)'] > 1)] \
                            .reset_index()['Report ID'].values
```

```python
# are the tasks longer?

print(h_a[h_a['Report ID'].isin(several_agencies_tasks)]['Task Turn-Around-Time (secs)'].mean())
print(h_a[h_a['Report ID'].isin(unique_agencies_tasks)]['Task Turn-Around-Time (secs)'].mean())


print(h_a[h_a['Report ID'].isin(several_agencies_tasks)]['Task Turn-Around-Time (secs)'].median())
print(h_a[h_a['Report ID'].isin(unique_agencies_tasks)]['Task Turn-Around-Time (secs)'].median())

print(h_a[h_a['Report ID'].isin(several_agencies_tasks)]['Task Turn-Around-Time (secs)'].std())
print(h_a[h_a['Report ID'].isin(unique_agencies_tasks)]['Task Turn-Around-Time (secs)'].std())
```

```python
# Median very close, but mean very different, because of a high std in the global report turnaround time
# it seems that reports that go onto several agencies are doing better

# i noticed some reports take an absurd amount of time and they seem to stay on the same agency
```

```python
# Let's see outliers

work_df = h_a[h_a['Task Type'] == 'Extraction'].copy()

Q1 = work_df['Task Turn-Around-Time (secs)'].quantile(0.02)
Q3 = work_df['Task Turn-Around-Time (secs)'].quantile(0.98)

IQR = Q3 - Q1
extraction_outliers = work_df[(work_df['Task Turn-Around-Time (secs)'] < (Q1 - 1.5 * IQR)) | 
              (work_df['Task Turn-Around-Time (secs)'] > (Q3 + 1.5 * IQR))]
```

```python
# Let's see outliers

work_df = h_a[h_a['Task Type'] == 'Fraud Assessment'].copy()

Q1 = work_df['Task Turn-Around-Time (secs)'].quantile(0.02)
Q3 = work_df['Task Turn-Around-Time (secs)'].quantile(0.98)

IQR = Q3 - Q1
fraud_outliers = work_df[(work_df['Task Turn-Around-Time (secs)'] < (Q1 - 1.5 * IQR)) | 
              (work_df['Task Turn-Around-Time (secs)'] > (Q3 + 1.5 * IQR))]
```

```python
# Let's see outliers

work_df = h_a[h_a['Task Type'] == 'Classification'].copy()

Q1 = work_df['Task Turn-Around-Time (secs)'].quantile(0.02)
Q3 = work_df['Task Turn-Around-Time (secs)'].quantile(0.98)

IQR = Q3 - Q1
classification_outliers = work_df[(work_df['Task Turn-Around-Time (secs)'] < (Q1 - 1.5 * IQR)) | 
              (work_df['Task Turn-Around-Time (secs)'] > (Q3 + 1.5 * IQR))]
```

```python
set_classif = set(classification_outliers['Report ID'].unique())
set_fraud = set(fraud_outliers['Report ID'].unique())
set_extract = set(extraction_outliers['Report ID'].unique())

print('Classification outliers : ', len(set_classif & set(unique_agencies_tasks))/len(set_classif), 'stayed same agency')
print('Fraud outliers : ', len(set_fraud & set(unique_agencies_tasks))/len(set_fraud), 'stayed same agency')
print('Extract outliers : ', len(set_extract & set(unique_agencies_tasks))/len(set_extract),  'stayed same agency')
```

```python
# Assumption: some agencies are good at dealing with certains tasks, especially for fraud assessment
# while for extraction and classification there even though they change, they still take long time
```

# hard task that are shared / hard task non shared 

```python
# we count for classification outliers the number of agencies and compare performance 

count_agencies_outliers = work_df[work_df['Report ID'].isin(classification_outliers['Report ID'].unique())]\
                    .groupby('Report ID')\
                    .agg({'Agency (Manual Processing Organisation)': 'nunique'})

several__agencies_outliers_reports = count_agencies_outliers[count_agencies_outliers['Agency (Manual Processing Organisation)'] > 1].index
unique__agencies_outliers_reports = count_agencies_outliers[count_agencies_outliers['Agency (Manual Processing Organisation)'] == 1].index

print(work_df[work_df['Report ID'].isin(several__agencies_outliers_reports)]\
              ['Report Turn-Around-Time (secs)'].mean())
print(work_df[work_df['Report ID'].isin(unique__agencies_outliers_reports)]\
              ['Report Turn-Around-Time (secs)'].mean())

# Goes with the idea that if tasks are too hard, they should be given to another agency that is more specialized.
# To get specialized agencies, we get for hard tasks the outliers and single agency ones


special_classif = work_df[work_df['Report ID'].isin(several__agencies_outliers_reports)]\
              .groupby('Agency (Manual Processing Organisation)').agg({'Task Turn-Around-Time (secs)': 'mean'})\
              .sort_values(by = 'Task Turn-Around-Time (secs)', ascending = True)
```

```python
# we count for extraction outliers the number of agencies and compare performance 

count_agencies_outliers = work_df[work_df['Report ID'].isin(extraction_outliers['Report ID'].unique())]\
                    .groupby('Report ID')\
                    .agg({'Agency (Manual Processing Organisation)': 'nunique'})

several__agencies_outliers_reports = count_agencies_outliers[count_agencies_outliers['Agency (Manual Processing Organisation)'] > 1].index
unique__agencies_outliers_reports = count_agencies_outliers[count_agencies_outliers['Agency (Manual Processing Organisation)'] == 1].index

print(work_df[work_df['Report ID'].isin(several__agencies_outliers_reports)]\
              ['Report Turn-Around-Time (secs)'].mean())
print(work_df[work_df['Report ID'].isin(unique__agencies_outliers_reports)]\
              ['Report Turn-Around-Time (secs)'].mean())


# Goes with the idea that if tasks are too hard, they should be given to another agency that is more specialized.
# To get specialized agencies, we get for hard tasks the outliers and single agency ones


special_extract = work_df[work_df['Report ID'].isin(several__agencies_outliers_reports)]\
              .groupby('Agency (Manual Processing Organisation)').agg({'Task Turn-Around-Time (secs)': 'mean'})\
              .sort_values(by = 'Task Turn-Around-Time (secs)', ascending = True)
```

```python
# we count for fraud outliers the number of agencies and compare performance 

count_agencies_outliers = work_df[work_df['Report ID'].isin(fraud_outliers['Report ID'].unique())]\
                    .groupby('Report ID')\
                    .agg({'Agency (Manual Processing Organisation)': 'nunique'})

several__agencies_outliers_reports = count_agencies_outliers[count_agencies_outliers['Agency (Manual Processing Organisation)'] > 1].index
unique__agencies_outliers_reports = count_agencies_outliers[count_agencies_outliers['Agency (Manual Processing Organisation)'] == 1].index

print(work_df[work_df['Report ID'].isin(several__agencies_outliers_reports)]\
              ['Report Turn-Around-Time (secs)'].mean())
print(work_df[work_df['Report ID'].isin(unique__agencies_outliers_reports)]\
              ['Report Turn-Around-Time (secs)'].mean())

# Goes with the idea that if tasks are too hard, they should be given to another agency that is more specialized.
# To get specialized agencies, we get for hard tasks the outliers and single agency ones

special_fraud = work_df[work_df['Report ID'].isin(several__agencies_outliers_reports)]\
              .groupby('Agency (Manual Processing Organisation)').agg({'Task Turn-Around-Time (secs)': 'mean'})\
              .sort_values(by = 'Task Turn-Around-Time (secs)', ascending = True)
```
